```python
import pandas as pd
import numpy as np
import geopandas as gpd
import rasterio

import os, h5py
from pathlib import Path
import openpyxl

import holoviews as hv
```

```python
lwi_struct_pt_loc = r'../../data/raw/Survey/West_Fork_12162022_RAS.shp'
```

```python
gdf = gpd.read_file(lwi_struct_pt_loc)
```

```python
gdf.head()
```

```python
bridge_codes = [str(x) for x in range(130,178)]
guardrail_codes = [str(x) for x in range(180,199)]
roadway_codes = [str(x) for x in range(47,95)]
cl_codes = [str(x) for x in range(47,54)]
drainagesruct_codes = [str(x) for x in range(250,297)]
```

```python
descriptions = gdf.DESCRIPTIO.unique().tolist()
```

 ### use cl codes to get gdf of centerline points in survey

```python
cl_descriptions = []
for i in descriptions:
    for j in cl_codes:
        if j in i:
            cl_descriptions.append(i)
```

```python
cl_gdf = gdf.loc[gdf['DESCRIPTIO'].isin(cl_descriptions)]
cl_gdf
```

```python
cl_gdf.to_file('../../data/interim/cl_points.shp')
```
